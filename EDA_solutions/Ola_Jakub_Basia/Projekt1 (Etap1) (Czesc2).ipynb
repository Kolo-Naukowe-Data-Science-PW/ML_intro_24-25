```python
# Zaimportowanie potrzebnych bibliotek
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis, skew, chi2_contingency
```

```python
# Wczytanie bazy danych
dataset = pd.read_csv('DATA.csv', encoding='utf-8')
dataset_raw = pd.read_csv('Loan_Default.csv', encoding='utf-8')
```

```python
# 3. Podstawowe statystyki opisowe
statistics = {}
numeric_columns = dataset_raw.select_dtypes(include=[np.number])
for column in numeric_columns:
    if column!="year":
        data = dataset_raw[column]
        stats = {
            'Arith mean': round(data.mean(),3),
            'Stand dev': round(data.std(),3),
            'Min': round(data.min(),3),
            'Max': round(data.max(),3),
            'Range': round(data.max() - data.min(),3),
            'Q1': round(data.quantile(0.25),3),
            'Q2 (med)': round(data.quantile(0.50),3),
            'Q3': round(data.quantile(0.75),3),
            'Kurtosis': round(kurtosis(data),3),
            'Skewness': round(skew(data),3)
            }
        statistics[column] = stats
statistics_df = pd.DataFrame(statistics).T
with open('3_podstawowe_statystyki.txt', 'w') as file: file.write(statistics_df.to_string())
```

```python
# 5.02/03. Obliczanie wspolczynnikow korelacji Spearmana i Pearsona

correlations = {}
numeric_columns = ['loan_amount', 'rate_of_interest', 'Interest_rate_spread', 'Upfront_charges', 'term', 'property_value',
                   'income', 'Credit_Score', 'age', 'LTV', 'dtir1', 'total_units']

for i in range(len(numeric_columns)):
    column1 = numeric_columns[i]
    for j in range(i+1, len(numeric_columns)):
        column2 = numeric_columns[j]
        pearson_corr = round(dataset[column1].corr(dataset[column2], method='pearson'), 3)
        spearman_corr = round(dataset[column1].corr(dataset[column2], method='spearman'), 3)
        correlations[f"{column1} vs {column2}"] = {
            'Pearson cor coeff': pearson_corr,
            'Spearman cor coeff': spearman_corr
        }
correlations_df = pd.DataFrame(correlations).T
correlations_df_sorted_P = correlations_df.sort_values(by='Pearson cor coeff', ascending=False)
correlations_df_sorted_S = correlations_df.sort_values(by='Spearman cor coeff', ascending=False)
with open('5_wspolczynniki_korelacji_P.txt', 'w') as file: file.write(correlations_df_sorted_P.to_string())
with open('5_wspolczynniki_korelacji_S.txt', 'w') as file: file.write(correlations_df_sorted_S.to_string())
```

```python
# 5.04. Wykonanie testu chi-kwadrat

chi2_results = {}
categorical_columns = ['loan_limit', 'Gender', 'approv_in_adv', 'loan_type', 'loan_purpose', 
                       'Credit_Worthiness', 'open_credit', 'business_or_commercial', 'Neg_ammortization',
                       'interest_only', 'lump_sum_payment', 'construction_type', 'occupancy_type', 'Secured_by',
                       'credit_type', 'co-applicant_credit_type', 'submission_of_application', 'Region', 
                       'Security_Type', 'Status']

for i in range(len(categorical_columns)):
    column1 = categorical_columns[i]
    for j in range(i+1, len(categorical_columns)):
        column2 = categorical_columns[j]
        contingency_table = pd.crosstab(dataset[column1], dataset[column2])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        chi2_results[f"{column1} vs {column2}"] = {
            'Chi2 Statistic': chi2,
            'p-value': p,
            'Degrees of Freedom': dof
        }
chi2_results_df = pd.DataFrame(chi2_results).T
chi2_results_df_sorted = chi2_results_df.sort_values(by='p-value', ascending=False)
with open('5_test_chi2.txt', 'w') as file: file.write(chi2_results_df_sorted.to_string())
```
