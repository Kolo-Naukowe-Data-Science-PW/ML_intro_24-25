```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Loan_Default.csv')
```

```python
sns.catplot(data=df, x='loan_type', kind='count', hue='Gender')
```

```python
data = df.copy()
cat_cols = [col for col in data.columns if data[col].dtype == 'object']
for col in cat_cols:
    unique_values = data[col].dropna().unique()
    unique_values_na = data[col].unique()
    value_mapping = {value: num for num, value in enumerate(unique_values)}
    data[col] = data[col].map(value_mapping)
    print(f"values for col {col} converted from {unique_values_na} to {data[col].unique()}")
corr_matrix = data.corr()
plt.figure(figsize=(30,20))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
dat = df.copy()
unique_values = dat['loan_limit'].unique()
unique_values_na = dat['loan_limit'].unique()
value_mapping = {value: num for num, value in enumerate(unique_values)}
dat['loan_limit'] = dat['loan_limit'].map(value_mapping)
print (f"values for col loan_limit converted from {unique_values_na} to {dat['loan_limit'].unique()}")
dat['loan_limit'] = dat['loan_limit'].replace(0, 'comforming loan')
dat['loan_limit'] = dat['loan_limit'].replace(1, 'NaN')
dat['loan_limit'] = dat['loan_limit'].replace(2, 'noncomforming loan')
sns.violinplot(data = dat, x='loan_limit', y = "loan_amount")
```
