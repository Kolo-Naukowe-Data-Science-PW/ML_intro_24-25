```python
# WybÃ³r zmiennych iloÅ›ciowych (tylko liczby)
numeric_columns = data.select_dtypes(include=['float64', 'int64']).drop(columns=['ID', 'year', 'Status'])

# Obliczenie macierzy korelacji
correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

## Wnioski
* SilnÄ… korelacje widaÄ‡ miÄ™dzy loan_amount a property_value, rÃ³wniez miÄ™dzy interest_rate_spread a rate_of_interest
* Ewentualnie mozna jeszcze rozwazyÄ‡ wyrÃ³zniajÄ…cÄ…Â korelacjÄ™Â miÄ™dzy income a loan_amount i miÄ™dzy income a property_value

```python
#Przygladam siÄ™ blizej zmiennym o najwiekszej korelacji

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='loan_amount', 
    y='property_value', 
    hue='Status',
    data=data, 
    alpha=0.5
)
plt.title('Loan Amount vs. property_value')
plt.xlabel('Loan Amount')
plt.ylabel('property_value')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Interest_rate_spread', 
    y='rate_of_interest', 
    hue='Status',
    data=data, 
    alpha=0.5
)
plt.title('Interest_rate_spread vs. rate_of_interest')
plt.xlabel('Interest_rate_spread')
plt.ylabel('rate_of_interest')
plt.show()
```
